```python
#import cv2
import cv2 as cv
import glob
import os
import numpy as np
from moviepy.editor import *
```

```python
#this fn seperate video to frames, set numOfFrame or sample_rate to None
def video2frame(video_path,frame_path,numOfFrame,sample_rate):
    file_list = os.listdir(video_path)
    i = 0
    for filename in file_list:
        if i == 10:
            break
        video = cv.VideoCapture(video_path + str(filename))
        
        os.mkdir(os.path.join(frame_path,os.path.join("positive"+str(i))))
        video_length = video.get(7)
        if video_length > 5000:
            continue
        if sample_rate == None:
            sample_rate = (video_length // numOfFrame)
            if sample_rate == 0:
                sample_rate = 1
        if numOfFrame == None:
            if video_length <= 700:
                sample_rate = int(sample_rate // 2)
            numOfFrame = int((video_length // sample_rate))
        print("the video length of %s is: %d"%(str(filename),video_length))
        
        k = 0
        while True:
            ret,image = video.read()
            if image is None:
                break
            if k % sample_rate == 0:
                image_name = frame_path + "positive" + str(i) + '/' + str(int(k//sample_rate)) + '.jpg'
                cv.imwrite(image_name,image)
            k += 1
            if (k / numOfFrame) == sample_rate:
                break
    
        i += 2
if __name__ == "__main__":
    video_path = "C:/Users/jingy/Jupyter/Feature/frames/Arson/Arson/"
    frame_path = "C:/Users/jingy/Jupyter/Fight/frames/Arson/"
    video2frame(video_path,frame_path,None,5)
```

```python
def video2frame_normal(video_path,frame_path,numOfFrame,sample_rate):
    file_list = os.listdir(video_path)
    i = 0
    for filename in file_list:
        video = cv.VideoCapture(video_path + str(filename))
        os.mkdir(os.path.join(frame_path,os.path.join("positive"+str(i))))
        video_length = video.get(7)
        if video_length > 5000:
            i += 1
            continue
        if sample_rate == None:
            sample_rate = (video_length // numOfFrame)
            if sample_rate == 0:
                sample_rate = 1
        if numOfFrame == None:
            if video_length <= 700:
                sample_rate = int(sample_rate // 2)
            numOfFrame = int((video_length // sample_rate))
        print("the video length of %s is: %d"%(str(filename),video_length))
        
        k = 0
        while True:
            ret,image = video.read()
            if image is None:
                break
            if k % sample_rate == 0:
                image_name = frame_path + "positive" + str(i) + '/' + str(int(k//sample_rate)) + '.jpg'
                cv.imwrite(image_name,image)
            k += 1
            if (k / numOfFrame) == sample_rate or k == 30 * sample_rate:
                break
    
        i += 1
if __name__ == "__main__":
    video_path = "E:/UCF-Anomaly-Detection-Dataset/UCF_Crimes/Videos/Training_Normal_Videos_Anomaly/"
    frame_path = "G:/anomaly/project/data/small_video/normal/"
    video2frame_normal(video_path,frame_path,None,5)
```

```python
#this fn will rename all the videos in the folder, the new video name will be: new_name+<video number>
def video_rename(path,new_name):
    count = 200
    file_list = os.listdir(path)
    for file in file_list:
        olddir = os.path.join(path,file)
        filename = os.path.splitext(file)[0]              
        filetype = os.path.splitext(file)[1]
        newdir = os.path.join(path+new_name+str(count)+filetype)
        os.rename(olddir,newdir)
        count += 1

path =  "G:/anomaly/project/data/small_video_1/normal/"
new_name = "negative"
video_rename(path,new_name)
```

```python
# this fn changes the fps of the video
def video_changeFPS(video_path,output_path):
    clip = VideoFileClip(video_path)
    clip.write_videofile(output_path, fps=120)
    clip.reader.close()

video_path = "C:/Users/18819/Videos/QQplayerClip/1.mp4"
output_path = "C:/Users/18819/Videos/QQplayerClip/111.mp4"
video_changeFPS(video_path,output_path)
```

```python
# frame to video
def frames2video(fps, save_path, frames_path):
    fourcc = cv.VideoWriter_fourcc(*'mp4v')
    videoWriter = cv.VideoWriter(save_path, fourcc, fps, (320, 240))
    for i in range(1000):
        if os.path.isfile("%s/%d.jpg"%(frames_path, i)):
            frame = cv.imread("%s/%d.jpg"%(frames_path, i))
            videoWriter.write(frame)
    videoWriter.release()
    return
 
frame_path = "G:/anomaly/project/data/small_video/normal/"
save_path ="G:/anomaly/project/data/small_video_1/normal/"
file_list = os.listdir(frame_path)
i = 0
for fn in file_list:
    print("prosessing: ",fn)
    frames2video(1,os.path.join(save_path+"negative"+str(i)+".mp4"),
                   os.path.join(frame_path+fn))
    i += 1
```
